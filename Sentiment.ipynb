```python
auth_id = []
degree_cen = []
with open('degreecentrality.txt') as infile:
    for line in infile:
        a = line.split()
        auth_id.append(a[0])
        degree_cen.append(a[1])

d = {'author_id' : auth_id,
     'degree_centrality' : degree_cen}

degree_cen = pd.DataFrame(d)
```

```python
def in_qrange(ser, q):
    return ser.between(*ser.quantile(q=q))
```

```python
degree_cen.degree_centrality = degree_cen.degree_centrality.astype(float)
```

```python
bottom25 =  degree_cen.loc[degree_cen['degree_centrality'].transform(in_qrange, q=[0, 0.25])]
top25 =  degree_cen.loc[degree_cen['degree_centrality'].transform(in_qrange, q=[0.75, 1])]
```
